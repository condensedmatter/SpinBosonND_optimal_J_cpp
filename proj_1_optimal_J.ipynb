```python
def run(datafolder,EachGroupSIZE,spinsize ):
     

    SEEDdigits=3
    DATAdigits=5
    
    
    LOOPS=10 
    sumN=20 

    %matplotlib inline
    import numpy as np
    import matplotlib
    import matplotlib.pyplot as plt

    def m_name(seed,i):
        return datafolder+str(seed).zfill(SEEDdigits)+"_"+str(i).zfill(DATAdigits)+".dat"

    def doit(seed,samplesize):
        d25=np.zeros((samplesize,spinsize))
        for i in range(samplesize):
            d25[i]=np.loadtxt(m_name(seed,i))
        d_m=np.mean(d25, axis=0)
        d_s=np.std(d25,axis=0)
        return np.median(np.log10(d_s/d_m/np.sqrt(samplesize)))

    x=[21, 34, 55, 89, 144, 233, 377, 610, 987, 1597, 2584 ]
    y=[doit(i,EachGroupSIZE) for i in x]

    timeSpend=np.loadtxt(datafolder+"timeSpend.txt")
    yt=np.log10(np.sqrt(timeSpend))

    plt.plot( x, y+yt[:len(x)],'-o')
    plt.title("-significant digits  Per second")
    plt.xlabel("J - jump steps")
    plt.ylabel("median of log10(relative error)")
    plt.show()
    
    plt.plot( x, 1/(timeSpend[:len(x)]/LOOPS/sumN/3600 ),'-o')
    plt.title("what sumN should it be, if want one file per hour")
    plt.xlabel("J - jump steps")
    plt.ylabel("sumN value")
    plt.show()
    
def cpuTime( SigPer1s):
    import numpy as np
    
    print( ["1 second"] +[ -SigPer1s  ] )
    print( ["1 hour"] +[ -SigPer1s-np.log10(np.sqrt(3600)) ] )
    print( ["1 day"] +[ -SigPer1s-np.log10(np.sqrt(3600*24)) ] )
    print( ["2 weeks"] +[ -SigPer1s-np.log10(np.sqrt(3600*24*14)) ] )
    print( ["3 day 32 cores"] +[ -SigPer1s-np.log10(np.sqrt(3600*24*3*32)) ] )
    print( ["1 day (2000 cores) "] +[ -SigPer1s-np.log10(np.sqrt(3600*24*2000)) ] )
    print( ["2 weeks (32 cores)x6"] +[ -SigPer1s-np.log10(np.sqrt(3600*24*14*32*6)) ] )
    print( ["2 weeks (2000 cores) "] +[ -SigPer1s-np.log10(np.sqrt(3600*24*2000*14)) ] )
    return None
```
